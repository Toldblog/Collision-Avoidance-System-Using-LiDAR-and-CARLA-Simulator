```python
import matplotlib.pyplot as plt

# Your color dictionary
color_values = {
    'red': [1, 0, 0], 'green': [0, 1, 0], 'blue': [0, 0, 1], 'yellow': [1, 1, 0],
    'orange': [1, 0.5, 0], 'purple': [0.5, 0, 0.5], 'cyan': [0, 1, 1], 'magenta': [1, 0, 1],
    'pink': [1, 0.75, 0.8], 'brown': [0.6, 0.3, 0.1], 'lime': [0.75, 1, 0], 'navy': [0, 0, 0.5],
    'olive': [0.5, 0.5, 0], 'teal': [0, 0.5, 0.5], 'maroon': [0.5, 0, 0], 'turquoise': [0.25, 0.88, 0.82],
    'gold': [1, 0.84, 0], 'silver': [0.75, 0.75, 0.75], 'indigo': [0.29, 0, 0.51], 'violet': [0.93, 0.51, 0.93]
}

# Determine grid size
num_colors = len(color_values)
columns = 5  # Number of columns in the grid
rows = (num_colors + columns - 1) // columns  # Calculate required rows

# Create a figure with subplots
fig, axes = plt.subplots(rows, columns, figsize=(15, rows * 3))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Iterate over colors and display them
for ax, (color_name, rgb_values) in zip(axes, color_values.items()):
    # Create a square filled with the color
    ax.add_patch(plt.Rectangle((0, 0), 1, 1, color=rgb_values))
    # Remove axes ticks and labels
    ax.axis('off')
    # Add the color name below the square
    ax.text(0.5, -0.1, color_name.capitalize(), ha='center', va='top', fontsize=12)

# Hide any empty subplots
for ax in axes[num_colors:]:
    ax.axis('off')

plt.tight_layout()
plt.show()
```
